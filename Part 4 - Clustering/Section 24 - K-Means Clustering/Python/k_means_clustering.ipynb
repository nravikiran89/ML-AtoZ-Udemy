# K-Means Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, 3:].values
```

## Using the elbow method to find the optimal number of clusters

```
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title("The elbow function")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

## Training the K-Means model on the dataset

```
kmeans = KMeans(n_clusters=7, init='k-means++', random_state=42)
kmeans.fit(X)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
```

## Visualising the clusters

```
plt.scatter(X[y_kmeans==0, 0], X[y_kmeans==0, 1], s=25, c='red', label='cluster 0')
plt.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], s=25, c='green', label='cluster 1')
plt.scatter(X[y_kmeans==2, 0], X[y_kmeans==2, 1], s=25, c='blue', label='cluster 2')
plt.scatter(X[y_kmeans==3, 0], X[y_kmeans==3, 1], s=25, c='black', label='cluster 3')
plt.scatter(X[y_kmeans==4, 0], X[y_kmeans==4, 1], s=25, c='magenta', label='cluster 4')
plt.scatter(X[y_kmeans==5, 0], X[y_kmeans==5, 1], s=25, c='grey', label='cluster 5')
plt.scatter(X[y_kmeans==6, 0], X[y_kmeans==6, 1], s=25, c='purple', label='cluster 6')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=50, c='cyan', label='Cluster Centers')

plt.title("K means cluster")
plt.xlabel("Annual income (k$)")
plt.ylabel("Spending score [1-100]")

plt.legend()
plt.show()
```
